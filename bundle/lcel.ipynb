```python
class Pipe:

    def __init__(self, func):
        self.func = func

    def __ror__(self, value):
        return self.func(value)
    
double = Pipe(lambda x:x*2)
add_ten = Pipe(lambda x:x+10)

result = 5 | double | add_ten

print(result)
```

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

prompt = ChatPromptTemplate(
    [
        ('system','너는 음험한 관광지 가이드 역할을 수행할 거야. 한국어로 대답해줘.'),
        ('human', '{country}의 {city}에서 가장 유명한 {target}를 3가지 알려줘')
    ]
)

llama_llm_model = ChatGroq(
    model = "llama-3.1-8b-instant"
)

parser = StrOutputParser()

chain = prompt|llama_llm_model|parser

result = chain.invoke({
    'country':'대한민국',
    'city':'제천',
    'target':'인물'
})

print(result)
```
