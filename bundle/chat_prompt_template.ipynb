```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()

groq_llm_model = ChatGroq(
    model = "llama-3.1-8b-instant"
)

chat_template_1 = ChatPromptTemplate.from_messages(
    [
        ('system', '너는 {language} 언어 선생님 역할을 수행할 거야'),
        ('human', '안녕, 오늘 배울 내용은 뭐야?'),
        ("ai", '안녕하세요!오늘은 문법 기초를 한번 배워볼까요?'),
        ('human', '{user_input}')
    ]
)

# msg_1 = chat_template_1.invoke({
#     'language':'여자',
#     'user_input':'숫자 세는 방법에 대해 설명해줘'
# })

result = groq_llm_model.invoke(chat_template_1.invoke({
    'language':'여자',
    'user_input':'숫자 세는 방법에 대해 설명해줘'
}))

print(result.content)
```
